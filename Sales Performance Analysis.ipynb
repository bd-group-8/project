# Sales Performance Overview
After a MQL filled a form on landing page to sign up for seller, a Sales Development Representative(SDR) contacted the MQL and gathered more information about the lead. Then a Sales Representative(SR) consulted the MQL. So interaction between SDRs/SRs and MQLs can affect conversion from MQLs to sellers.

At this section we will deal with two aspects of sales result, conversion rate and sales length.

Note: A MQL who finally signed up for seller is called a _closed deal_.

```
# Merge 'mql' with 'cd'
# Merge by 'left' in order to evaluate conversion rate

mql_cd = mql.join(cd, mql.mql_id == cd.mql_id,how='left').drop(cd.mql_id)

display(mql_cd)
```

```
print(mql_cd.count(), "rows") # Row Count
print(len(mql_cd.columns), "Columns") # Column Count
```

```
# Add a column to distinguish signed MOLs from MQLs who left without signing up
mql_cd_1 = mql_cd.withColumn("seller_id_bool", mql_cd.seller_id.isNotNull())
display(mql_cd_1)
```

```
# Add a 'year-month' column
import re
from pyspark.sql.functions import udf
@udf
def date_Y_M(date):
  #print(text)
  return re.search('\d{4}-\d{2}', str(date)).group()

mql_cd_2 = mql_cd_1.withColumn("first_contact_date_Y_M", date_Y_M(mql_cd_1.first_contact_date))
display(mql_cd_2)
```

```
# Another way to Add a 'year-month' column
#from pyspark.sql.functions import expr
#mql_cd_2 = mql_cd_1.withColumn("first_contact_date_Y_M", expr("to_date(first_contact_date, 'yyyy-MM')"))
#display(mql_cd_2)
```

```
# Using sqlContext.sql or spark.sql to write SQL instead of sql.functions

mql_cd_2.createOrReplaceTempView("mql_cd_2")

monthly_cd = spark.sql("select first_contact_date_Y_M, count(case when seller_id_bool=True then 1 end) as number_of_closed_deals from mql_cd_2 \
group by first_contact_date_Y_M \
order by first_contact_date_Y_M")
display(monthly_cd)
```

#### Monthly volume of closed deals sharply increased in 2018.

## Conversion Rate
Conversion rate means the percentage of MQLs who finally signed up for sellers (closed deals).

```
# Calculate monthly conversion rate
monthly_conversion = spark.sql("select first_contact_date_Y_M, \
count(case when seller_id_bool=True then 1 end) as number_of_closed_deals, \
count(first_contact_date_Y_M) as number_of_mqls, \
round(count(case when seller_id_bool=True then 1 end)/count(first_contact_date_Y_M)*100, 1) as conversion_rate from mql_cd_2 \
group by first_contact_date_Y_M \
order by first_contact_date_Y_M")
display(monthly_conversion)
```

#### Monthly conversion rate increased with volume.

## Sales Length
Sales length means period from first contact to signing up for seller.

```
from pyspark.sql.functions import datediff, to_date

mql_cd_3 = mql_cd_2.withColumn("sales_length",datediff(to_date(mql_cd_2.won_date, 'yyyy-MM-dd'),to_date(mql_cd_2.first_contact_date, 'yyyy-MM-dd')))
mql_cd_3.createOrReplaceTempView("mql_cd_3")
display(mql_cd_3)
```

```
# Count closed deals by sales length for year 2017
sales_length_2017 = spark.sql("select sales_length, \
count(case when seller_id_bool=True then 1 end) as number_of_closed_deals from mql_cd_3 \
where first_contact_date_Y_M like '2017%' \
group by sales_length \
order by sales_length")
sales_length_2017.createOrReplaceTempView("sales_length_2017")
display(sales_length_2017)
```

```
# Count closed deals by sales length for year 2018
sales_length_2018 = spark.sql("select sales_length, \
count(case when seller_id_bool=True then 1 end) as number_of_closed_deals from mql_cd_3 \
where first_contact_date_Y_M like '2018%' \
group by sales_length \
order by sales_length")
sales_length_2018.createOrReplaceTempView("sales_length_2018")
display(sales_length_2018)
```

```
# Calculate running total of closed deals by sales length for year 2017
# NOT efficient replace it later
sales_length_2017_running_total = spark.sql("select a.sales_length, \
a.number_of_closed_deals, \
sum(b.number_of_closed_deals) from sales_length_2017 as a \
join sales_length_2017 as b \
on (a.sales_length >= b.sales_length) \
group by a.sales_length, a.number_of_closed_deals \
order by a.sales_length")
display(sales_length_2017_running_total)
```

```
# Calculate running total of closed deals by sales length for year 2018
# NOT efficient replace it later
sales_length_2018_running_total = spark.sql("select a.sales_length, \
a.number_of_closed_deals, \
sum(b.number_of_closed_deals) from sales_length_2018 as a \
join sales_length_2018 as b \
on (a.sales_length >= b.sales_length) \
group by a.sales_length, a.number_of_closed_deals \
order by a.sales_length")
display(sales_length_2018_running_total)
```

##### Sales length was dramatically shortened in 2018 as well.
While few deals 8 out of 61 (around 13%) were closed within 50 days in 2017, 616 out of 781 (around 79%) were closed within 50 days in 2018.
##### Generally, both conversion rate and sales length were improved in 2018 against 2017.
